```python
import bs4
import urllib
import urllib3
import pandas as pd
```

```python
main_links = ['https://www.ricardoanaya.com.mx/sala-de-prensa']
for page in range(2,16):
    main_links.append('https://www.ricardoanaya.com.mx/sala-de-prensa/page/'+ str(page))
```

```python
def make_soup(myurl):
    if 'â€œ' not in myurl:
        pm = urllib3.PoolManager()
        html = pm.urlopen(url = myurl, method = 'GET').data
        return bs4.BeautifulSoup(html, 'html5lib')
```

```python
def get_anaya(page, results_list):
    for link in page.find_all('a', rel= "Index, Follow"):
        results_list.append('https://www.ricardoanaya.com.mx' + link.get('href'))
```

```python
all_links = []
```

```python
for url in main_links:
    soup = make_soup(url)
    get_anaya(soup, all_links)
```

```python
def get_text(soup_page, delete = ['\n', '\xa0']):
    if soup_page.find('div',style="float: left; width: 100%; margin-bottom: 15px;"):
        my_string = soup.find('div', style="float: left; width: 100%; margin-bottom: 15px;").text
        for sign in delete:
            my_string = my_string.replace(sign,'')
        my_string = my_string.replace('\t',' ')
        return my_string
```

```python
def get_date(soup_page):
    if soup_page:
        if soup_page.find('div',style="float: left; width: 100%; margin-bottom: 15px;"):
            date = soup_page.find('div', style="float: left; width: 100%; margin-bottom: 15px;").find_all('div')
            if len(date) > 1:
                return date[1].text
            else:
                return ''
```

```python
dates = []
for link in all_links:
    soup = make_soup(link)
    dates.append(get_date(soup))
```

```python
len(dates)
```

```python
anaya_texts = []
for link in all_links:
    link_soup = make_soup(link)
    date = get_date(link_soup)
    text = get_text(link_soup)
    anaya_texts.append([date, text])
```

```python
anaya_df = pd.DataFrame(anaya_texts, columns=['date', 'text'])
```
