```python
import bs4
import requests
import urllib3
import nltk
import pandas as pd
```

```python
main_links = ['https://lopezobrador.org.mx/secciones/boletines/']
for page in range(2,15):
    main_links.append('https://lopezobrador.org.mx/secciones/boletines/page/'+ str(page) + '/')
```

```python
def make_soup(myurl):
    pm = urllib3.PoolManager()
    html = pm.urlopen(url = myurl, method = 'GET').data
    return bs4.BeautifulSoup(html, 'html5lib')
```

```python
def get_text(boletin_soup):
    initial = ''
    for i in range(3,len(boletin_soup.find_all('p', style="text-align: justify;"))):
        initial += boletin_soup.find_all('p', style="text-align: justify;")[i].text
    return initial
```

```python
def get_date(boletin):
    return boletin.next_sibling.text
```

```python
def get_boletines(main_link, texts_list):
    soup = make_soup(main_link)
    boletines = soup.find_all("h2", class_="entry-title")
    for boletin in boletines:
        subsoup = boletin.next_sibling.next_sibling
        texts_list.append(get_text(subsoup))
```

```python
def get_boletines_date(main_link, texts_list):
    soup = make_soup(main_link)
    boletines = soup.find_all("h2", class_="entry-title")
    for boletin in boletines:
        subsoup = boletin.next_sibling.next_sibling
        if get_text(subsoup) != '':
            texts_list.append([get_date(boletin), get_text(subsoup)])
```

```python
amlo_texts = []
for main_link in main_links:
    get_boletines_date(main_link, amlo_texts)
```

```python
amlo_df = pd.DataFrame(amlo_texts, columns=['date', 'text'])
```
