```python
import bs4
import urllib
import urllib3
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
```

```python
main_links = ['https://www.ricardoanaya.com.mx/sala-de-prensa']
for page in range(2,16):
    main_links.append('https://www.ricardoanaya.com.mx/sala-de-prensa/page/'+ str(page))
```

```python
def make_soup(myurl):
    if '“' not in myurl:
        pm = urllib3.PoolManager()
        html = pm.urlopen(url = myurl, method = 'GET').data
        return bs4.BeautifulSoup(html, 'html5lib')
```

```python
def get_anaya(page, results_list):
    for link in page.find_all('a', rel= "Index, Follow"):
        results_list.append('https://www.ricardoanaya.com.mx' + link.get('href'))
```

```python
all_links = []
```

```python
for url in main_links:
    soup = make_soup(url)
    get_anaya(soup, all_links)
```

```python
def clean_str(string):
    string = string.lower()
    stop_words = stopwords.words('spanish') + ['sala', 'prensa']
    if string in stop_words:
        return ''
    else:
        return string
```

```python
def clean_text(long_string):
    tokenizer = RegexpTokenizer(r'\w+')
    str_list = tokenizer.tokenize(long_string)
    new = ' '
    final_list = [clean_str(word) for word in str_list]
    return new.join(final_list)
```

```python
def get_text(soup_page, delete = ['\n', '\xa0']):
    if soup_page.find('div',style="float: left; width: 100%; margin-bottom: 15px;"):
        my_string = soup.find('div', style="float: left; width: 100%; margin-bottom: 15px;").text
        for sign in delete:
            my_string = my_string.replace(sign,'')
        my_string = my_string.replace('\t',' ')
        return clean_text(my_string)
```

```python
def get_anaya_text(soup, delete = ['\n', '\xa0', '\t', ',', '.', ';', ':', '"', '-', '_', '(', ')', '[', ']', '{', '}', '¿', '?', '¡', '!', '*', '“', '”', '•']):
    if soup.find('div', style="float: left; width: 100%; margin-bottom: 15px;"):
        my_string = soup.find('div', style="float: left; width: 100%; margin-bottom: 15px;").text
        for sign in delete:
            my_string = my_string.replace(sign,'')
        return my_string
```

```python
def get_date(soup_page):
    if soup_page:
        if soup_page.find('div',style="float: left; width: 100%; margin-bottom: 15px;"):
            date = soup_page.find('div', style="float: left; width: 100%; margin-bottom: 15px;").find_all('div')
            if len(date) > 1:
                return date[1].text
            else:
                return ''
```

```python
def write_txt(text, file_name):
    with open(file_name, 'w') as f:
        f.write(text)
```

```python
anaya_text = []
```

```python
for link in all_links:
    soup = make_soup(link)
    if soup:
        anaya_text.append(get_text(soup))
```

```python
for i in range(len(anaya_text)):
    file_name = 'anaya_txts/anaya_' + str(i) + '.txt'
    if anaya_text[i] != None:
        write_txt(anaya_text[i], file_name)
```

```python
anaya_df = pd.DataFrame(anaya_text, columns = ['text'])
```
