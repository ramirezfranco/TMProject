```python
import bs4
import requests
import urllib3
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
import pandas as pd
```

```python
main_links = ['https://lopezobrador.org.mx/secciones/boletines/']
for page in range(2,15):
    main_links.append('https://lopezobrador.org.mx/secciones/boletines/page/'+ str(page) + '/')
```

```python
def make_soup(myurl):
    pm = urllib3.PoolManager()
    html = pm.urlopen(url = myurl, method = 'GET').data
    return bs4.BeautifulSoup(html, 'html5lib')
```

```python
def get_text(boletin_soup):
    initial = ''
    for i in range(3,len(boletin_soup.find_all('p', style="text-align: justify;"))):
        initial += boletin_soup.find_all('p', style="text-align: justify;")[i].text
    return initial
```

```python
def get_date(boletin):
    return boletin.next_sibling.text
```

```python
def clean_str(string):
    string = string.lower()
    for i in ["ª","'", "\n", "\t", ",", "\xa0"]:
        string = string.replace(i, '')
    if string in stopwords.words('spanish'):
        return ''
    else:
        return string
```

```python
def clean_text(long_string):
    tokenizer = RegexpTokenizer(r'\w+')
    str_list = tokenizer.tokenize(long_string)
    new = ' '
    final_list = [clean_str(word) for word in str_list]
    return new.join(final_list)
```

```python
def get_boletines(main_link, texts_list):
    soup = make_soup(main_link)
    boletines = soup.find_all("h2", class_="entry-title")
    for boletin in boletines:
        subsoup = boletin.next_sibling.next_sibling
        texts_list.append(get_text(subsoup))
```

```python
def get_boletines_date(main_link, texts_list):
    soup = make_soup(main_link)
    boletines = soup.find_all("h2", class_="entry-title")
    for boletin in boletines:
        subsoup = boletin.next_sibling.next_sibling
        if get_text(subsoup) != '':
            texts_list.append([get_date(boletin), clean_text(get_text(subsoup))])
```

```python
def write_txt(text, file_name):
    with open(file_name, 'w') as f:
        f.write(text)
```

```python
amlo_texts = []
for main_link in main_links:
    get_boletines_date(main_link, amlo_texts)
```

```python
amlo_df = pd.DataFrame(amlo_texts, columns=['date', 'text'])
```

```python
amlo_nov = amlo_df[:162]
```

```python
for i in range(len(amlo_texts)):
    file_name = 'amlo_txts/amlo_' + str(i) + '.txt'
    write_txt(amlo_texts[i][1], file_name)
```

```python
all_links = ['https://lopezobrador.org.mx/secciones/boletines/']
for page in range(2,112):
    all_links.append('https://lopezobrador.org.mx/secciones/boletines/page/'+ str(page) + '/')
```
